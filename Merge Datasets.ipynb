```python
import pandas as pd
import os
```

```python
def merge(folder_path):
    csv_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path)
                 if os.path.isfile(os.path.join(folder_path, f)) and f.endswith('.csv')]
    print(len(csv_files))
    print(csv_files)
    merged = pd.DataFrame([])
    for record in csv_files:
        df = pd.read_csv(record)
        merged = merged.append(df)
    return merged
```

```python
tmp = merge("tables/amazon/digital_music/tuning_general/")
```

```python
deduplicated_tmp = tmp.drop_duplicates()
```

```python
file_index = 1

for file_index in range(1, 3):
    file_name = 'tables/amazon/digital_music/tuning_general/merged/mmp-part' + str(file_index) + '.csv'
    deduplicated_tmp.to_csv(file_name, index=False)
```
