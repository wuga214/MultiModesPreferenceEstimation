```python
import pandas as pd
import os
```

```python
def find_best_hyperparameters(folder_path):
    csv_files = [os.path.join(folder_path, f) for f in os.listdir(folder_path)
                 if os.path.isfile(os.path.join(folder_path, f)) and f.endswith('.csv')]
    merged = pd.DataFrame([])
    for record in csv_files:
        df = pd.read_csv(record)
        merged = merged.append(df)
    return merged
```

```python
tmp = find_best_hyperparameters("tables/netflix/tuning_general/")
```

```python
file_index = 1

for file_index in range(1, 217):
    file_name = 'tables/netflix/tuning_general/merged/mmp-part' + str(file_index) + '.csv'
    tmp.to_csv(file_name, index=False)
```
